```python
def get_first_digit(x):
    while x >= 10:
        x /= 10
    return int(x)
def get_sec_digit(x):
    while x >= 100:
        x /= 10
    return int(x%10)
def get_first_two_digits(x):
    while x >= 100:
        x /= 10
    return int(x)
```

```python
# get frequencies for 1st digit and 1st 2 digit Benford's law distr
def get_benford_pdf(lst):
    probs = []
    for d in lst:
        probs.append(math.log(1+1/d,10))
    return probs
```

```python
first_digits = np.arange(1,10)
first_dig_pdf = get_benford_pdf(first_digits)
first_dig_cdf = np.cumsum(first_dig_pdf)

first_two_digits = np.arange(10,100)
first_two_dig_pdf = get_benford_pdf(first_two_digits)
first_two_dig_cdf = np.cumsum(first_two_dig_pdf)
```

```python
# TESTS

def stat_tests(data,law_type,exp_pdf):
    # law type= 0 if single digit BL, 1 if two digits BL
    
    digits = np.arange(10**law_type,10**law_type + len(exp_pdf))
    # get pdf and cdf of data
    data.sort()
    counter = collections.Counter(data)
    total = len(data)
    
    data_count = [0]*len(exp_pdf)
    for i in counter:
        data_count[i-10**law_type] = counter[i]
    data_pdf = np.divide(data_count,total)
    data_cdf = np.cumsum(data_pdf)
    exp_cdf = np.cumsum(exp_pdf)
    exp_count = [round(p * total) for p in exp_pdf]
    
    # plot data with expected distribution
    plt.plot(digits,exp_pdf,marker=".")
    plt.bar(digits,data_pdf,color='0.75')
    plt.xlabel("Digit")
    plt.ylabel("Relative Frequency")
    plt.show()
    
    # chi square test (8 degrees of freedom & P-val=0.05)
    chi_square_stat = 0  # chi square test statistic
    for data, expected in zip(data_count,exp_count):

        chi_square = math.pow(data - expected, 2)

        chi_square_stat += chi_square / expected

    print("\nChi-squared Test Statistic = {:.3f}".format(chi_square_stat))
    print("Critical value at a P-value of 0.05 is 15.51.")    

    # kolmogorov smirnov test
    ks_stat = np.max(np.abs(np.subtract(exp_cdf, data_cdf)))
    ks_crit_val = 1.36/(total**0.5)
    print("KS Test Statistic = {:.3f}".format(ks_stat))
    print("Critical value at a P-value of 0.05 is {:.3f}".format(ks_crit_val))  
    
    # kuiper test
    Dplus = np.max(np.subtract(exp_cdf, data_cdf))
    Dminus = np.max(np.subtract(data_cdf, exp_cdf))
    k_stat = Dplus + Dminus
    print("Kuiper Test Statistic = {:.3f}".format(k_stat))
    print("Critical value at a P-value of 0.05 is 1.321") 
    
    # cramer von mise statistics
    p = exp_pdf + [exp_pdf[0]]
    t = []
    for i in range(len(exp_pdf)):
        t.append((p[i]+p[i+1])/2)
        
    Z = np.subtract(data_cdf,exp_cdf)
    Z_bar = np.sum(np.multiply(Z,t))
    
    cvm_vals = np.multiply(np.power(Z,2),t)
    cvm_stat = total * np.sum(cvm_vals)
    
    watson_vals = np.multiply(np.power(np.subtract(Z,Z_bar),2),t)
    watson = total * np.sum(watson_vals)
    
    ad_vals = np.divide(np.multiply(np.power(Z,2),t),np.multiply(exp_pdf,np.subtract(1,exp_pdf)))
    ad_stat = total * np.sum(ad_vals)
    
    return [chi_square_stat,ks_stat,k_stat,cvm_stat,watson,ad_stat]
```

```python
# evaluating benford's law for china city first digits
stat_tests(china_city_first,0,first_dig_pdf)
```

```python
stat_tests(china_city_firsttwo,1,first_two_dig_pdf)
```
